#1. import libraries

```python
import pandas as pd #data manipulation
import numpy as np #mathematical operation
import matplotlib.pyplot as plt #data visualization
from sklearn.linear_model import LinearRegression #Data modeling//Training data
```

#2. Create DataFrame and Saving to CSV

```python
df = pd.read_csv('Lesson_4_student_grades_dataset.csv')

print(df.head)
```

#3. Visualizing the data

```python
plt.scatter(df['Hours_of_Study'], df['Grades'])
plt.xlabel('attendance percentage')
plt.ylabel('previous exam scores')
plt.title("Scatter plot of Data")
plt.show()
```

#4. preparing data for modeling

```python
x = df[['Attendance_Percentage']]
y = df[['Previous_Exam_Scores']]
```

5. fitting the linear regression model

```python
model = LinearRegression()
model.fit(x,y)
```

#6. Getting model Coefficients

```python
slope = model.coef_[0]
intercept = model.intercept_
print("slope:  ",slope)
print("intercept: ", intercept)
```

7. Visualizing the linear line

```python
plt.scatter(x,y)
plt.plot(x,model.predict(x),color='blue')
plt.xlabel('independent variable(X)')
plt.ylabel('independent variable(Y)')
plt.title("Scatter plot of Data")
plt.show()
```

#8. making predictions (extra)

```python
new_x = [[np.random.randint(50,100)]]
predicted_y = model.predict(new_x)
print("Value of X: ", new_x)
print("Predicted Y for new X: ", predicted_y)
```
